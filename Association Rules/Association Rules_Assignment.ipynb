```python
# Drop rows with missing values
df.isnull().sum()
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Convert each row into a list of items
transactions = df.iloc[:, 0].apply(lambda x: x.split(',')).tolist()
transactions[:5]
```

```python
# Transaction Encoding
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
basket_df = pd.DataFrame(te_ary, columns=te.columns_)
basket_df.head()
```

```python
# Apply Apriori
frequent_itemsets = apriori(basket_df, min_support=0.02, use_colnames=True)
frequent_itemsets.sort_values(by='support', ascending=False).head()
```

```python
# Generate Association Rules
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1)
rules_sorted = rules.sort_values(by='lift', ascending=False)
rules_sorted[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```

```python
# 1. What is lift and why is it important in Association rules?
```

```python
# Definition:
# Lift measures how much more likely two items are to be bought together than if they were independent.

# Lift= Confidence(A→B)/Support(B)
 
# Interpretation:
# Lift = 1: A and B are independent.
# Lift > 1: A and B are positively correlated (more likely to occur together).
# Lift < 1: A and B are negatively correlated (less likely to occur together).

# Importance:
# Lift helps identify truly interesting rules. It filters out rules that may seem strong (e.g., high confidence) but actually occur just 
# because the consequent is very common.
```

```python
# 2. What is support and Confidence. How do you calculate them?
```

```python
# Support:
# Support tells how frequently the itemset appears in the dataset.

# Support(𝐴→𝐵) = Transactions containing both A and B / Total transactions 
 
# Confidence:
# Confidence indicates how often B is bought when A is bought.

# Confidence(𝐴→𝐵) = Support(𝐴∪𝐵) / Support(𝐴)
```

```python
# 3. What are some limitations or challenges of Association rules mining?
```

```python
# | **Challenge**     | **Explanation**                                                                |
# | ----------------- | ------------------------------------------------------------------------------ |
# | Too Many Rules    | Can generate thousands of rules, many of which are uninteresting.              |
# | Rare Item Problem | Rare but valuable associations may be missed if support threshold is too high. |
# | Scalability       | Large datasets can make computation slow and memory intensive.                 |
# | Lack of Context   | Association rules only show correlation, not causation.                        |
# | Redundancy        | Many rules might express the same pattern in slightly different ways.          |
```
