```python
# 2. Model Implementation
```

```python
# ●	Construct a basic ANN model using your chosen high-level neural network library. 
# Ensure your model includes at least one hidden layer.
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = df.drop('letter', axis='columns')
y = df.letter	
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
y_onehot = tf.keras.utils.to_categorical(y_encoded)
```

```python
y_onehot
```

```python
# ●	Divide the dataset into training and test sets.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y_onehot, test_size=0.2, random_state=42)
```

```python
# ●	Train your model on the training set and then use it to make predictions on the test set.
```

```python
model = Sequential()
model.add(Dense(64, input_shape=(x.shape[1],), activation='relu'))  # Hidden layer
model.add(Dense(32, activation='relu'))  # second hidden layer
model.add(Dense(y_onehot.shape[1], activation='softmax'))  # Output layer
```

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, epochs=20, batch_size=32, validation_split=0.2)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f"\nTest Accuracy: {accuracy:.4f}")
```
