```python
# â€¢	Discuss the pros and cons of One-Hot Encoding and Label Encoding.
```

```python
# 1. Label Encoding

# What it does:
# Converts each unique category in a column into a unique integer value.
# Example:
# education = [â€˜Highschoolâ€™, â€˜Bachelorsâ€™, â€˜Mastersâ€™] â†’
# [0, 1, 2]

# Pros:
# Simple and fast to apply.
# Doesnâ€™t increase the dimensionality of the data.
# Useful when categorical variables are ordinal (i.e., order matters: low < medium < high).

# Cons:
# Can mislead ML algorithms if categories are nominal (no order) â€” they may treat them as ordinal.
# Impacts models like Linear Regression or KNN, which may interpret label values as having magnitude or distance.
```

```python
# 2. One-Hot Encoding

# What it does:
# Creates new binary (0/1) columns for each category.
# Example:
# education = [â€˜Highschoolâ€™, â€˜Bachelorsâ€™, â€˜Mastersâ€™] â†’
# education_Highschool, education_Bachelors, education_Masters

# Pros:
# Ideal for nominal categorical variables (no inherent order).
# Prevents the model from assuming any sort of order or priority between categories.
# Works well with algorithms like Logistic Regression, Random Forests, Neural Networks.

# Cons:
# Increases the dimensionality of the dataset (can lead to curse of dimensionality).
# Less efficient when a variable has many categories (e.g., native-country in Adult dataset).
# Some ML models can struggle with high-dimensional sparse data.
```

```python
# How Outliers Affect Model Performance:

# Distorted Feature Distributions: Outliers can skew the distribution of features, which may confuse models like Linear Regression or Logistic Regression.
# Misleading Loss Functions: Many models optimize based on minimizing error (like Mean Squared Error). Outliers heavily influence such metrics.
# Decision Boundaries: For tree-based models or SVMs, outliers can shift decision boundaries, leading to poor generalization.
# Longer Training Time: Complex models may require more time and resources when outliers are present due to noise.
# Lower Accuracy: Ultimately, predictive performance can degrade as the model tries to fit both the regular patterns and the noise introduced by outliers.
```
